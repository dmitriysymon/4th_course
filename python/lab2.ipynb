# Завдання №2

```python
import random

# Список ключів
keys = ['GDP', 'Inflation', 'Unemployment', 'Consumer Confidence', 'Interest Rates']
n = len(keys)
print(n, keys)

# Параметри нормального розподілу для кожного ключа
means = [3, 100, 10, 75, 16]
stddevs = [5, 30, 5, 15, 3]

# Створення порожнього словника
economic_data = {}

# Генерація випадкових значень та додавання їх до словника
for i in range(n):
    values = [round(random.normalvariate(means[i], stddevs[i]), 2) for _ in range(10)]
    values.sort()  # Сортування списку значень
    economic_data[keys[i]] = values

# Виведення списків як текстові рядки
for key, values in economic_data.items():
    values_str = ', '.join(map(str, values))
    print(f"{key}: {values_str}")
```

# Завдання №3

```python
import math

# Розрахунок параметрів описової статистики для кожного ключа
for key in economic_data.keys():
    values = economic_data[key]
    n = len(values)

    # Розрахувати мінімальне та максимальне значення
    min_val = min(values)
    max_val = max(values)

    # Розрахувати середнє значення
    mean = sum(values) / n

    # Розрахувати медіану значень
    sorted_values = sorted(values)
    if n % 2 == 0:
        median = (sorted_values[n // 2 - 1] + sorted_values[n // 2]) / 2
    else:
        median = sorted_values[n // 2]

    # Розрахувати перший квартиль (25-й перцентиль) та третій квартиль (75-й перцентиль)
    q1_index = int(n * 0.25)
    q3_index = int(n * 0.75)
    q1 = sorted_values[q1_index]
    q3 = sorted_values[q3_index]

    # Розрахувати стандартне відхилення значень
    s = math.sqrt(sum([(x - mean) ** 2 for x in values]) / (n - 1))

    # Вивести параметри описової статистики на екран
    print(f"Описова статистика для {key}:")
    print(f"Мінімальне значення: {min_val}")
    print(f"Максимальне значення: {max_val}")
    print(f"Сума значень: {sum(values)}")
    print(f"Кількість значень: {n}")
    print(f"Середнє значення: {mean}")
    print(f"Медіана: {median}")
    print(f"Перший квартиль (Q1): {q1}")
    print(f"Третій квартиль (Q3): {q3}")
    print(f"Стандартне відхилення: {s}")
    print()
```

# Завдання №5

```python
import csv
from statistics import StatisticsError, mean, median, mode, variance, stdev, quantiles

# Створення списку словників на основі вихідних даних
list2 = [dict(zip(economic_data, t)) for t in zip(*economic_data.values())]
fieldnames = list(list2[0].keys())

# Створення нового файлу
with open('economic_data.csv', 'w') as f:
    fieldnames = list(list2[0].keys())
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for row in list2:
        writer.writerow(row)

# Додавання даних до файлу
new_row = {'GDP': -2.11,
           'Inflation': 267.47,
           'Unemployment': 18.7,
           'Consumer Confidence': 205.29,
           'Interest Rates': 22.72,
           'Positive GDP Growth': False}
with open('economic_data.csv', 'a') as f:
    writer = csv.DictWriter(f, fieldnames=list(new_row.keys()))
    writer.writerow(new_row)

# Зчитування даних з файлу
with open('economic_data.csv', 'r') as f:
    reader = csv.DictReader(f)
    data = [row for row in reader]

# Створення масиву даних
data_dict = {}

data_dict['GDP'] = [round(random.normalvariate(1000, 100), 2) for i in range(30)]  # мільярди USD
data_dict['Population'] = [round(random.normalvariate(50, 10), 2) for i in range(30)]  # мільярди осіб
data_dict['CPI'] = [round(random.normalvariate(100, 10), 2) for i in range(30)]  # відсотки (геометрична середня)

# Функція для обчислення описової статистики
def calculate_statistics(data):
    # Середнє значення
    avg = mean(data)
    print(f"Середнє значення: {avg}")

    # Медіана
    med = median(data)
    print(f"Медіана: {med}")

    # Мода
    try:
        mod = mode(data)
        print(f"Мода: {mod}")
    except StatisticsError:
        print("Немає моди, всі значення унікальні.")

    # Дисперсія
    var = variance(data)
    print(f"Дисперсія: {var}")

    # Стандартне відхилення
    std = stdev(data)
    print(f"Стандартне відхилення: {std}")

    # Квартилі
    quartiles_list = quantiles(data, n=4)
    q1, q2, q3 = quartiles_list[0], quartiles_list[1], quartiles_list[2]
    print(f"Перший квартиль (Q1): {q1}")
    print(f"Другий квартиль (Q2): {q2}")
    print(f"Третій квартиль (Q3): {q3}")

# Виведення описової статистики для кожного ключа в масиві даних
for key, values in data_dict.items():
    print(f"Описова статистика для '{key}':")
    calculate_statistics(values)
    print()
```
